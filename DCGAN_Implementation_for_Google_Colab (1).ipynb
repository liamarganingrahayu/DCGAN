```
# ==============================================================================
# DCGAN (Deep Convolutional Generative Adversarial Network) Implementation
# Didesain untuk Google Colab dengan TensorFlow 2.x
#
# Pakar: Advanced Deep Learning Research & Python Colab Coding
# Dataset: MNIST (untuk demonstrasi pembuatan gambar)
# ==============================================================================

import tensorflow as tf
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
import os
import time

# --- 1. Konfigurasi Lingkungan dan Hyperparameter ---

# Pastikan GPU terdeteksi dan digunakan
if tf.config.list_physical_devices('GPU'):
    print("GPU Terdeteksi dan Akan Digunakan.")
else:
    print("Perhatian: GPU tidak terdeteksi. Pelatihan mungkin sangat lambat.")

# Hyperparameters
BUFFER_SIZE = 60000 # Ukuran buffer untuk shuffling dataset
BATCH_SIZE = 256    # Ukuran batch untuk pelatihan
EPOCHS = 50         # Jumlah epoch yang akan dijalankan (dapat disesuaikan)
noise_dim = 100     # Dimensi vektor input noise (latents space)
num_examples_to_generate = 16 # Jumlah gambar yang akan dihasilkan untuk visualisasi

# Seed untuk menghasilkan gambar yang konsisten
seed = tf.random.normal([num_examples_to_generate, noise_dim])

# Folder untuk menyimpan output (gambar yang dihasilkan)
checkpoint_dir = './training_checkpoints'
checkpoint_prefix = os.path.join(checkpoint_dir, "ckpt")

# --- 2. Persiapan Data (MNIST) ---

# Muat dan siapkan dataset MNIST
(train_images, train_labels), (_, _) = tf.keras.datasets.mnist.load_data()

# Reshape dan normalisasi:
# 1. Tambahkan dimensi channel (28x28 -> 28x28x1)
train_images = train_images.reshape(train_images.shape[0], 28, 28, 1).astype('float32')

# 2. Normalisasi ke [-1, 1] seperti yang umum digunakan dalam GAN
# Pixel value (0-255) -> (0-1) -> (-1, 1)
train_images = (train_images - 127.5) / 127.5  # Skala ke [-1, 1]

# Buat objek tf.data.Dataset
train_dataset = tf.data.Dataset.from_tensor_slices(train_images).shuffle(BUFFER_SIZE).batch(BATCH_SIZE)

print(f"Bentuk data pelatihan setelah preprocessing: {train_images.shape}")
print(f"Jumlah batch per epoch: {len(train_dataset)}")


# --- 3. Definisi Model DCGAN ---

# 3.1. Generator Model (Model Penghasil Gambar)
# Mengambil noise_dim (100) dan menghasilkan gambar 28x28x1
def make_generator_model():
    model = tf.keras.Sequential(name='Generator')

    # Dimensi input noise (100)
    # Dense layer untuk mengubah noise menjadi representasi feature map 7x7x256
    model.add(layers.Dense(7*7*256, use_bias=False, input_shape=(noise_dim,)))
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    # Reshape output ke 7x7x256
    model.add(layers.Reshape((7, 7, 256)))
    assert model.output_shape == (None, 7, 7, 256) # Batch size tidak ada dalam None

    # Layer 1: Conv2DTranspose (Upsampling ke 14x14)
    model.add(layers.Conv2DTranspose(128, (5, 5), strides=(1, 1), padding='same', use_bias=False))
    assert model.output_shape == (None, 7, 7, 128)
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    # Layer 2: Conv2DTranspose (Upsampling ke 14x14)
    model.add(layers.Conv2DTranspose(64, (5, 5), strides=(2, 2), padding='same', use_bias=False))
    assert model.output_shape == (None, 14, 14, 64)
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    # Layer 3: Conv2DTranspose (Upsampling ke 28x28)
    # Output layer menggunakan tanh untuk menghasilkan output dalam rentang [-1, 1]
    model.add(layers.Conv2DTranspose(1, (5, 5), strides=(2, 2), padding='same', use_bias=False, activation='tanh'))
    assert model.output_shape == (None, 28, 28, 1)

    return model

# 3.2. Discriminator Model (Model Pembeda/Pengklasifikasi)
# Mengambil gambar 28x28x1 dan menghasilkan prediksi real/fake (1/0)
def make_discriminator_model():
    model = tf.keras.Sequential(name='Discriminator')

    # Layer 1: Conv2D
    model.add(layers.Conv2D(64, (5, 5), strides=(2, 2), padding='same',
                                     input_shape=[28, 28, 1]))
    model.add(layers.LeakyReLU())
    model.add(layers.Dropout(0.3))

    # Layer 2: Conv2D
    model.add(layers.Conv2D(128, (5, 5), strides=(2, 2), padding='same'))
    model.add(layers.LeakyReLU())
    model.add(layers.Dropout(0.3))

    # Flatten dan output
    model.add(layers.Flatten())
    # Output tunggal (probabilitas Real/Fake)
    model.add(layers.Dense(1))

    return model

# Inisialisasi model
generator = make_generator_model()
discriminator = make_discriminator_model()

# --- 4. Fungsi Loss dan Optimizer ---

# Loss function: Binary Crossentropy (umum untuk masalah klasifikasi biner)
cross_entropy = tf.keras.losses.BinaryCrossentropy(from_logits=True)

# 4.1. Discriminator Loss
# Mengukur seberapa baik discriminator membedakan antara gambar asli (label 1) dan palsu (label 0)
def discriminator_loss(real_output, fake_output):
    # Loss untuk gambar asli (seharusnya 1)
    real_loss = cross_entropy(tf.ones_like(real_output), real_output)
    # Loss untuk gambar palsu (seharusnya 0)
    fake_loss = cross_entropy(tf.zeros_like(fake_output), fake_output)
    # Total loss
    total_loss = real_loss + fake_loss
    return total_loss

# 4.2. Generator Loss
# Mengukur seberapa baik generator menipu discriminator (seharusnya memprediksi 1)
def generator_loss(fake_output):
    # Generator ingin discriminator memprediksi 1 (real) untuk gambar palsu
    return cross_entropy(tf.ones_like(fake_output), fake_output)

# 4.3. Optimizers
# Adam Optimizer dengan learning rate yang berbeda (seringkali membantu stabilitas GAN)
generator_optimizer = tf.keras.optimizers.Adam(learning_rate=1e-4, beta_1=0.5)
discriminator_optimizer = tf.keras.optimizers.Adam(learning_rate=1e-4, beta_1=0.5)

# --- 5. Checkpointing (Penyimpanan Model) ---

# Setup Checkpoint manager untuk menyimpan dan me-restore model
checkpoint = tf.train.Checkpoint(generator_optimizer=generator_optimizer,
                                 discriminator_optimizer=discriminator_optimizer,
                                 generator=generator,
                                 discriminator=discriminator)


# --- 6. Fungsi Pelatihan (@tf.function) ---

# Dekorator @tf.function mengkompilasi fungsi menjadi graph TensorFlow yang cepat
@tf.function
def train_step(images):
    # 1. Noise Generator
    noise = tf.random.normal([BATCH_SIZE, noise_dim])

    # Persistent=True agar gradient tape dapat mengakses variabel untuk G dan D
    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
        # 2. Generator menghasilkan gambar palsu
        generated_images = generator(noise, training=True)

        # 3. Discriminator memproses gambar asli dan palsu
        real_output = discriminator(images, training=True)
        fake_output = discriminator(generated_images, training=True)

        # 4. Hitung Loss
        gen_loss = generator_loss(fake_output)
        disc_loss = discriminator_loss(real_output, fake_output)

    # 5. Hitung Gradient
    gradients_of_generator = gen_tape.gradient(gen_loss, generator.trainable_variables)
    gradients_of_discriminator = disc_tape.gradient(disc_loss, discriminator.trainable_variables)

    # 6. Aplikasikan Gradient
    generator_optimizer.apply_gradients(zip(gradients_of_generator, generator.trainable_variables))
    discriminator_optimizer.apply_gradients(zip(gradients_of_discriminator, discriminator.trainable_variables))

    return gen_loss, disc_loss

# --- 7. Fungsi Visualisasi dan Logging ---

# Fungsi untuk menghasilkan dan memplot gambar dari Generator
def generate_and_save_images(model, epoch, test_input):
    # 'training' disetel False agar Batch Normalization berjalan dalam mode inferensi
    predictions = model(test_input, training=False)

    fig = plt.figure(figsize=(4, 4))
    plt.suptitle(f"Epoch {epoch}", fontsize=16)

    for i in range(predictions.shape[0]):
        plt.subplot(4, 4, i+1)
        # Denormalisasi gambar dari [-1, 1] ke [0, 1] untuk plot
        # Gunakan cmap='gray' untuk gambar grayscale
        plt.imshow(predictions[i, :, :, 0] * 0.5 + 0.5, cmap='gray')
        plt.axis('off')

    # Simpan gambar di Colab
    plt.savefig(f'image_at_epoch_{epoch:04d}.png')
    plt.show()

# --- 8. Fungsi Pelatihan Utama ---

def train(dataset, epochs):
    history = {'G_loss': [], 'D_loss': []}

    print("\n--- Memulai Pelatihan DCGAN ---")
    for epoch in range(epochs):
        start = time.time()

        g_loss_epoch = 0
        d_loss_epoch = 0
        batch_count = 0

        for image_batch in dataset:
            g_loss, d_loss = train_step(image_batch)
            g_loss_epoch += g_loss.numpy()
            d_loss_epoch += d_loss.numpy()
            batch_count += 1

        # Rata-rata Loss untuk epoch
        avg_g_loss = g_loss_epoch / batch_count
        avg_d_loss = d_loss_epoch / batch_count
        history['G_loss'].append(avg_g_loss)
        history['D_loss'].append(avg_d_loss)

        # Logging dan Visualisasi
        print (f'Epoch {epoch + 1:04d}, '
               f'G Loss: {avg_g_loss:.4f}, '
               f'D Loss: {avg_d_loss:.4f}, '
               f'Time: {time.time()-start:.2f} sec')

        # Hasilkan gambar setiap 5 epoch (atau setiap epoch jika ingin melihat perkembangannya)
        if (epoch + 1) % 5 == 0 or epoch == epochs - 1:
            generate_and_save_images(generator, epoch + 1, seed)
            checkpoint.save(file_prefix = checkpoint_prefix)

    # Hasilkan satu set gambar terakhir
    generate_and_save_images(generator, epochs, seed)
    return history

# --- 9. Eksekusi Pelatihan ---

# Memastikan folder checkpoint ada
os.makedirs(checkpoint_dir, exist_ok=True)

# Memuat checkpoint terbaru jika ada
latest_checkpoint = tf.train.latest_checkpoint(checkpoint_dir)
if latest_checkpoint:
    checkpoint.restore(latest_checkpoint)
    start_epoch = int(latest_checkpoint.split('-')[-1]) * 5 # Estimasi
    print(f"Melanjutkan dari checkpoint terbaru: {latest_checkpoint}")
else:
    start_epoch = 0
    print("Memulai pelatihan dari awal.")

# Pelatihan model
loss_history = train(train_dataset, EPOCHS)

# --- 10. Visualisasi History Loss ---

plt.figure(figsize=(10, 6))
plt.plot(loss_history['G_loss'], label='Generator Loss')
plt.plot(loss_history['D_loss'], label='Discriminator Loss')
plt.title('DCGAN Training Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss Value')
plt.legend()
plt.grid(True)
plt.show()

# --- 11. Generasi Gambar Akhir (Opsional) ---

# Generasi 25 gambar baru dari noise acak setelah pelatihan selesai
random_noise = tf.random.normal([25, noise_dim])
final_images = generator(random_noise, training=False)

fig = plt.figure(figsize=(5, 5))
plt.suptitle("Generated Images After Training", fontsize=16)

for i in range(final_images.shape[0]):
    plt.subplot(5, 5, i+1)
    # Denormalisasi gambar dan plot
    plt.imshow(final_images[i, :, :, 0] * 0.5 + 0.5, cmap='gray')
    plt.axis('off')

plt.show()

print("\n--- Pelatihan DCGAN Selesai ---")
print("Cek folder './training_checkpoints' untuk model yang disimpan.")
print("Cek output untuk melihat perkembangan gambar dari epoch ke epoch.")
```
